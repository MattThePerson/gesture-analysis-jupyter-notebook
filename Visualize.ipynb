```python
import pandas
from matplotlib import pyplot as plt

paikallaan = "paikallaan.csv"
nykaisy = "nykaisy.csv"
nosto = "nosto.csv"
custom = "custom.csv"
petting = "petting.csv"

df = pandas.read_csv(petting)

print(df)
```

```python
# plot acc data
ax = plt.gca()

df.plot(x="time", y="ax",ax=ax)
df.plot(x="time", y="ay",ax=ax)
df.plot(x="time", y="az",ax=ax)

plt.title("accelerometer data")
plt.xlabel("time (ms)")
plt.ylabel("acceleration (g)")
plt.legend()
plt.show()

# plot gyro data
ax = plt.gca()

df.plot(x="time", y="gx",ax=ax)
df.plot(x="time", y="gy",ax=ax)
df.plot(x="time", y="gz",ax=ax)

plt.title("gyroscope data")
plt.xlabel("time (ms)")
plt.ylabel("angular velocity (deg/s)")
plt.legend()
plt.show()
```
